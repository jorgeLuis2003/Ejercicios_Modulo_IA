```python
import numpy as no
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
entrenamiento = pd.read_csv('Hipertension_Arterial_Mexico.csv')
```

```python
data = pd.read_csv('Hipertension_Arterial_Mexico.csv')
```

```python
data
```

```python
entrenamiento.drop('FOLIO_I',axis=1,inplace=True)
```

```python
x=entrenamiento.drop('riesgo_hipertension',axis=1)
y=entrenamiento['riesgo_hipertension']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=30)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
modelo = Sequential()
```

```python
modelo.add(Dense(34,activation='relu',input_shape=(x_train_scaled.shape[1],)))
modelo.add(Dense(34,activation='relu'))
modelo.add(Dense(34,activation='relu'))
#34 numero de conexiones siempre deben ser 3
```

```python
x_test
```

```python
modelo.add(Dense(1,activation='sigmoid'))

```

```python
modelo.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) #para ver que tan preciso es el binario
```

```python
#entrenamiento

curva=modelo.fit(x_train,y_train,epochs=30,batch_size=32,validation_split=0.1) #las epocas es la cantidad de veces que va a leer mientras mas epocas tenga un modelo aprendera mejorbatch_size=32, #implica que cada 32 datos va a validation_split=0.1 # es un coeficiente de validacion
    
```

```python
plt.xlabel('# de epocas')
plt.ylabel('# magnitud de perdida')
plt.plot(curva.history['loss'])
```

```python
prediccion=modelo.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,roc_auc_score
```

```python
mse=mean_squared_error(y_test,prediccion)
mae=mean_absolute_error(y_test,prediccion)
r2=r2_score(y_test,prediccion)
auc_roc = roc_auc_score(y_test,prediccion)
```
